```python
# Gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x='Plataforma', y='Ventas NA', data=df_graficar)
plt.title('Gráfico de Barras')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

```python
# Gráfico de torta (para columnas categóricas)
conteo = df['Genero'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(conteo, labels=conteo.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Gráfico de Torta')
plt.show()
```

```python
# Gráfico de torta (para columnas categóricas)
conteo = df_graficar['Editorial'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(conteo, labels=conteo.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Gráfico de Torta')
plt.show()
```

```python
correlacion = df.corr()
sns.heatmap(correlacion, annot=True, cmap='coolwarm')
plt.show()
```

```python
columnas_numericas = df_graficar.select_dtypes(include=['int'])
for col in columnas_numericas:
    sns.histplot(data=df_graficar, x='Año', kde=False)
    plt.title(f'Histograma por Año')
    plt.show()
```

```python
top_80 = df.head(80)
print(top_80)
df_graficar2= top_80
df_graficar2.head(80)
```

```python
ventas_genero = df_graficar2.groupby('Genero')['Editorial'].value_counts().unstack()
ventas_genero.plot.line(marker='o')
plt.xlabel('Genero')
plt.ylabel('Número de Editoriales')
plt.title('Número de Editoriales por Género a lo largo de los años')
plt.legend(loc='best')
plt.show()
```

```python
plt.scatter(df['Ventas NA'], df['Ventas EU'])
plt.xlabel('Ventas Norteamérica')
plt.ylabel('Ventas Europa')
plt.title('Ventas de Videojuegos en Norteamérica y Europa')
plt.show()
```

```python
ventas_plataforma = df_graficar.groupby('Plataforma')['Ventas Global'].sum()
ventas_plataforma.plot.pie(autopct='%1.1f%%')
plt.axis('equal')
plt.title('Ventas Globales de Videojuegos por Plataforma')
plt.legend(loc='best')
plt.show()
```

```python
df.groupby('Año')['Ventas Global'].sum().plot.bar()
plt.xlabel('Año')
plt.ylabel('Ventas Globales')
plt.title('Ventas Globales de Videojuegos por Año')
plt.show()
```
